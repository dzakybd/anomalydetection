## Load library

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import OneClassSVM
from sklearn.ensemble import IsolationForest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.metrics import confusion_matrix, precision_recall_curve
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import PCA
from scipy import stats
from scipy.signal import find_peaks
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras import regularizers
from sklearn.linear_model import LogisticRegression
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn import preprocessing 
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
import seaborn as sns
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, f1_score, accuracy_score, recall_score, precision_score, auc
from pickle import dump
from numpy.random import seed
import tensorflow
from sklearn.model_selection import ShuffleSplit
from numpy import quantile

rseed = 1
seed(rseed)
tensorflow.random.set_seed(rseed)
sns.set(style="whitegrid")
pd.options.display.max_columns = None
```

# Data preprocessing

```python
data = pd.read_csv('creditcard.csv') # Load data
data = data.drop(columns=['Time']) # drop unused variable 'Time'
print(data.isnull().values.any()) # Check if the data has missing value

data.head()
```

### Normalization

```python
scaled = preprocessing.MinMaxScaler().fit_transform(data.values)
data_scale = pd.DataFrame(scaled, columns=data.columns, index=data.index)
data_scale
```

```python
print(data_scale["Class"].value_counts()) #Class distribution
data_scale['Class'].value_counts().plot(kind='bar', title="Original Class Distribution")
```

### Undersample

```python
y = data_scale["Class"].values
X_normal, X_fraud = data_scale[y == 0].iloc[:,:-1], data_scale[y == 1].iloc[:,:-1]
Y_normal, Y_fraud = np.zeros(X_normal.shape[0]), np.ones(X_fraud.shape[0])
outlier_fraction = len(Y_fraud)/float(len(Y_normal))
print("Outlier fraction before:", outlier_fraction)
print("Fraud/Valid samples : {} / {}".format(len(Y_fraud), len(Y_normal)))

# undersample normal data -> from 99.83% & 0.17% to 90% to 10%
X_normal = X_normal.sample(n=len(Y_fraud)*9, random_state=rseed)
X_normal = X_normal.reset_index(drop=True)
Y_normal = np.zeros(X_normal.shape[0])

outlier_fraction = len(Y_fraud)/float(len(Y_normal))
print("Outlier fraction after:", outlier_fraction)
print("Fraud/Valid samples : {} / {}".format(len(Y_fraud), len(Y_normal)))

#Class distribution
temp = pd.DataFrame(np.concatenate((Y_normal, Y_fraud)), columns=['Class'])
temp['Class'].value_counts().plot(kind='bar', title="Class Distribution after Undersampling")
```

### Data splitting

```python
# Splitting Train & Test of Normal data
X_train_normal, X_test_normal, Y_train_normal, Y_test_normal = train_test_split(X_normal, Y_normal, test_size=0.2)
# Splitting Train & Test of Fraud data
X_train_fraud, X_test_fraud, Y_train_fraud, Y_test_fraud = train_test_split(X_fraud, Y_fraud, test_size=0.2)

# Combine Train Normal & Fraud data
X_train_all = np.append(X_train_normal, X_train_fraud, axis = 0)
Y_train_all = np.append(Y_train_normal, Y_train_fraud).astype(int)

# Combine Test Normal & Fraud data
X_test_all = np.append(X_test_normal, X_test_fraud, axis = 0)
Y_test_all = np.append(Y_test_normal, Y_test_fraud).astype(int)
```

# Model building & Evaluation

```python
# GridSearch setting
cv = ShuffleSplit(test_size=0.2, n_splits=1, random_state=rseed)

def evaluation_matrix(Y_test_all, y_pred, title):
    print(classification_report(Y_test_all,y_pred))
    print("accuracy_score", accuracy_score(Y_test_all,y_pred))
    print("f1_score", f1_score(Y_test_all,y_pred,average=None))
    print("recall_score", recall_score(Y_test_all,y_pred,average=None))
    print("precision_score", precision_score(Y_test_all,y_pred,average=None))
    
    matrix = confusion_matrix(Y_test_all, y_pred)
    labels = ['Normal', 'Fraud']
    g=sns.heatmap(matrix,annot=True, cbar=False, fmt='g', xticklabels=labels, yticklabels=labels)
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.title('Confusion matrix of {}'.format(title))
    plt.tight_layout()
    plt.savefig("cf_{}.png".format(title))
    plt.show()
    plt.close()
    
def dimensionality_plot(X, y, title):
    pc = PCA(n_components=2, random_state = rseed)
    # Fitting the data
    X_trans = pc.fit_transform(X)
    plt.figure()
    plt.scatter(X_trans[np.where(y == 0), 0], X_trans[np.where(y==0), 1], marker='o', color='g', linewidth='1', alpha=0.8, label='Normal')
    plt.scatter(X_trans[np.where(y == 1), 0], X_trans[np.where(y==1), 1], marker='o', color='k', linewidth='1', alpha=0.8, label='Fraud')
    plt.legend(loc = 'upper right')
    plt.title('2D Plot {}'.format(title))
    plt.tight_layout()
    plt.savefig("2dplot_{}.png".format(title))
    plt.show()
    plt.close()
```

## Isolation forest (unsupervised)

```python
param_grid = {'n_estimators': [100, 200],
            'max_samples': ['auto', 100],
            'contamination': [outlier_fraction, 'auto']}
f1_ms = make_scorer(f1_score, average='binary', pos_label=-1)

clf_if = IsolationForest(random_state=rseed)
clf_if = GridSearchCV(clf_if, param_grid, cv=cv, scoring=f1_ms, verbose=100, refit=True)

# Provide class label with sklearn format
Y_train_all_2 = Y_train_all.copy()
Y_train_all_2[Y_train_all_2 == 1] = -1
Y_train_all_2[Y_train_all_2 == 0] = 1

clf_if.fit(X_train_all, Y_train_all_2)
```

```python
# Gridsearch Parameter Tuning result
print(clf_if.best_params_)
print(clf_if.best_score_)
dump(clf_if.best_estimator_, open('isolation_forest.pkl', 'wb'))
```

```python
y_pred = clf_if.best_estimator_.predict(X_test_all)
y_pred[y_pred == 1] = 0
y_pred[y_pred == -1] = 1

evaluation_matrix(Y_test_all, y_pred, "Isolation Forest")
```

## Random forest (supervised)

```python
param_grid = {'n_estimators': [50, 100],
              'criterion': ['gini', 'entropy'],
             'min_samples_split': [2, 4]}

f1_ms = make_scorer(f1_score, average='binary', pos_label=1)

clf_rf = RandomForestClassifier(random_state=rseed)
clf_rf = GridSearchCV(clf_rf, param_grid, cv=cv, scoring=f1_ms, verbose=100, refit=True)
clf_rf.fit(X_train_all, Y_train_all)
```

```python
# Gridsearch Parameter Tuning result
print(clf_rf.best_params_)
print(clf_rf.best_score_)
dump(clf_rf.best_estimator_, open('random_forest.pkl', 'wb'))
```

```python
y_pred = clf_rf.best_estimator_.predict(X_test_all)
evaluation_matrix(Y_test_all, y_pred, "Random Forest")
```

## One-class SVM (semi-supervised)

```python
param_grid = {'kernel': ['rbf'],
             'gamma': ['scale', 0.1, 1],
             'nu': [0.1, 0.5, 1]}
f1_ms = make_scorer(f1_score)

clf_svm = OneClassSVM()
clf_svm = GridSearchCV(clf_svm, param_grid, cv=cv, scoring=f1_ms, verbose=100, refit=True)

Y_train_normal_2 = Y_train_normal.copy()
Y_train_normal_2[Y_train_normal_2 == 1] = -1
Y_train_normal_2[Y_train_normal_2 == 0] = 1

clf_svm.fit(X_train_normal, Y_train_normal_2)
```

```python
# Gridsearch Parameter Tuning result
print(clf_svm.best_params_)
print(clf_svm.best_score_)
dump(clf_svm.best_estimator_, open('oneclass_svm.pkl', 'wb'))
```

```python
y_pred = clf_svm.best_estimator_.predict(X_test_all)
y_pred[y_pred == 1] = 0
y_pred[y_pred == -1] = 1

evaluation_matrix(Y_test_all, y_pred, "One-Class SVM")
```

## One-class Autoencoder
1. Hybrid (semi-supervised & supervised): encoder + classifier
1. Semi-supervised: using threshold

```python
def create_model(encoded_neuron=32, depth=3):
    input_layer = Input(shape=(X_train_normal.shape[1],))
    encoded = input_layer
    
    ## encoding part
    if depth==5:
        encoded = Dense(encoded_neuron*16)(encoded)
    if depth==4:
        encoded = Dense(encoded_neuron*8)(encoded)
    encoded = Dense(encoded_neuron*4)(encoded)
    encoded = Dense(encoded_neuron*2)(encoded)
    encoded = Dense(encoded_neuron)(encoded)

    ## decoding part
    decoded = Dense(encoded_neuron)(encoded)
    decoded = Dense(encoded_neuron*2)(decoded)
    decoded = Dense(encoded_neuron*4)(decoded)
    if depth==4:
        decoded = Dense(encoded_neuron*8)(decoded)
    if depth==5:
        decoded = Dense(encoded_neuron*16)(decoded)
        
    ## output layer
    output_layer = Dense(X_train_normal.shape[1], activation='sigmoid')(decoded)

    autoencoder = Model(input_layer, output_layer)
    autoencoder.compile(optimizer='adam', loss="mse")
    
    return autoencoder
```

```python
params = {'encoded_neuron': [8, 16, 32],
         'depth': [3, 4, 5]}

autoencoder = KerasRegressor(build_fn=create_model)

autoencoder_grid = GridSearchCV(autoencoder, param_grid=params, cv=cv, verbose=100, scoring='neg_mean_squared_error', refit=True) 

autoencoder_grid.fit(X_train_normal, X_train_normal, shuffle=True, epochs=300)
```

```python
# Gridsearch Parameter Tuning result
print(autoencoder_grid.best_params_)
print(autoencoder_grid.best_score_)
autoencoder_grid.best_estimator_.model.save('oneclass_ae.h5')
```

### Semi-supervised: error threshold-based

```python
# Get reconstruction error of train normal data
pred = autoencoder_grid.best_estimator_.model.predict(X_train_normal)
mse = np.mean(np.power(X_train_normal - pred, 2), axis=1)

error_train_normal = pd.DataFrame({'reconstruction_error': mse, 'true_class': Y_train_normal})
error_train_normal.head()
```

```python
# Get reconstruction error of train data
pred = autoencoder_grid.best_estimator_.model.predict(X_test_all)
mse = np.mean(np.power(X_test_all - pred, 2), axis=1)

error_test = pd.DataFrame({'reconstruction_error': mse, 'true_class': Y_test_all})
error_test.head()
```

```python
error_test_normal = error_test.iloc[:len(Y_test_all)-len(Y_test_fraud),0]
error_test_fraud = error_test.iloc[len(Y_test_all)-len(Y_test_fraud):,0]

max_etn = max(error_test_normal)
min_etn = min(error_test_normal)
min_etf = min(error_test_fraud)
print(max_etn)
print(max_etn)
print(min_etf)
if min_etn <= min_etf <= max_etn:
    print("Yes")
```

```python
err_normal_sorted = sorted(error_test_normal)
err_normal_sorted = pd.Series(err_normal_sorted)
err_normal_sorted.plot()
```

```python
err_test_sorted = sorted(error_test_fraud)
err_test_sorted = pd.Series(err_test_sorted)
err_test_sorted.plot()
```

```python
# Get threshold with 99.5 confidence interval
threshold = quantile(error_train_normal['reconstruction_error'].values, 0.995)
print('{0:.10f}'.format(threshold))
```

```python
groups = error_test.groupby('true_class')
fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group.index, group.reconstruction_error, marker='o', ms=3.5, linestyle='',
            label= "Fraud" if name == 1 else "Normal")
ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.legend()
plt.title("Sample reconstruction error")
plt.ylabel("MSE error")
plt.xlabel("Sample index")
plt.tight_layout()
plt.savefig("sample_error.png")
plt.show()
plt.close()
```

```python
# prediction
y_pred = [1 if e > threshold else 0 for e in error_test.reconstruction_error.values]
evaluation_matrix(error_test.true_class, y_pred, "One-Class Autoencoder")
```
